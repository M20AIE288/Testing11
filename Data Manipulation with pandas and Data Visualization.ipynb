### Correlation:Â df.corr() -  calculates the correlation coefficients between all pairs of numeric columns in the DataFrame and prints the correlation matrix. Each cell in the correlation matrix represents the correlation coefficient between the corresponding pair of columns. The correlation coefficient ranges from -1 to 1, where:

- 1 indicates a perfect positive correlation,
- -1 indicates a perfect negative correlation, and
- 0 indicates no correlation.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Sample data
data = {
    'A': [1, 2, 3, 4, 5],
    'B': [5, 4, 3, 2, 1],
    'C': [2, 3, 4, 5, 6],
    'D': [3, 2, 5, 4, 1]
}

# Create DataFrame
df = pd.DataFrame(data)

# Calculate correlation matrix
correlation_matrix = df.corr()

# Print correlation matrix
print("Correlation Matrix:")
print(correlation_matrix)

# Plot correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Matrix')
plt.show()

```

### This command generates a box plot for the column named 'numeric_column' in the DataFrame. A box plot visually represents the distribution of numerical data and helps identify outliers, if any.

### Deletion 

### Remove rows with missing values

```python
import pandas as pd

# Sample DataFrame with missing values
data = {'A': [1, 2, None, 4, 5],
        'B': [None, 6, 7, 8, 9],
        'C': [10, 11, 12, None, 14]}

df = pd.DataFrame(data)

print("Original DataFrame:")
print(df)

# Remove rows with missing values
df.dropna(inplace=True)
print("\nDataFrame after removing rows with missing values:")
print(df)
```

### Imputation

### Forward or backward Fill missing values with the mean


### Forward Fill - In the forward fill method, missing values are filled with the last observed non-null value in each column.

```python
import pandas as pd
import numpy as np

# Create sample DataFrame with missing values
data = {'A': [1, np.nan, 3, np.nan, 5],
        'B': [np.nan, 2, np.nan, 4, np.nan]}
df = pd.DataFrame(data)

# Forward fill missing values
df_ffill = df.ffill()
print("DataFrame after forward fill:")
print(df_ffill)
```

### Backward Fill - In the backward fill method, missing values are filled with the next observed non-null value in each column.

```python
import pandas as pd
import numpy as np

# Create sample DataFrame with missing values
data = {'A': [1, np.nan, 3, np.nan, 5],
        'B': [np.nan, 2, np.nan, 4, np.nan]}
df = pd.DataFrame(data)

# Backward fill missing values
df_bfill = df.bfill()
print("\nDataFrame after backward fill:")
print(df_bfill)
```

### Indicator Variables


### In the context of missing data, indicator variables can be used to flag whether a certain value is missing or not. This can be helpful for analysis and modeling, as missing data often needs to be handled or accounted for in some way.

### This code creates a new column named 'column_name_missing' in the DataFrame. The purpose of this column is to indicate whether the corresponding value in the column 'column_name' is missing or not.

```python
import pandas as pd

# Sample data
data = {'column_name': [10, 20, None, 30, None, 40]}
df = pd.DataFrame(data)

# Indicator Variables
df['column_name_missing'] = df['column_name'].isnull().astype(int)

print(df)
```

# Creating Basic Plot


### Line plot 

A line plot is a type of graph that displays data points connected by straight lines. It is commonly used to visualize the relationship between two variables over a continuous interval.

```python
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]
y = [10, 14, 8, 20, 12]

# Create a line plot
plt.plot(x, y, marker='o', linestyle='-', color='b', label='Line Plot')

# Adding labels and a title
plt.xlabel('X-axis')  # Label for the X-axis
plt.ylabel('Y-axis')  # Label for the Y-axis
plt.title('Line Plot Example')  # Title of the plot

# Adding a legend
plt.legend()  # Show the legend with label specified in the plot

# Show the plot
plt.show()  # Display the plot
```

### Scatter plot

A scatter plot is a type of data visualization that displays the relationship between two variables by representing data points as individual dots on a two-dimensional plane

```python
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]
y = [10, 14, 8, 20, 12]

# Create a scatter plot
plt.scatter(x, y, marker='o', color='r', label='Scatter Plot')

# Adding labels and a legend
plt.xlabel('X-axis')  # Label for the X-axis
plt.ylabel('Y-axis')  # Label for the Y-axis
plt.title('Scatter Plot Example')  # Title of the plot
plt.legend()  # Show legend

# Show the plot
plt.show()
```

### Bar Chart

A bar chart is a type of graph that presents categorical data with rectangular bars. The length or height of each bar is proportional to the value it represents. 

```python
import matplotlib.pyplot as plt

# Sample data
categories = ['Category A', 'Category B', 'Category C', 'Category D']
values = [15, 24, 20, 30]

# Create a bar chart
plt.bar(categories, values, color='g', alpha=0.7, label='Bar Chart')

# Adding labels and a legend
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Chart Example')
plt.legend()

# Show the plot
plt.show()
```

# Seaborn

### Seaborn is a data visualization library.
 - Built on top of Matplotlib.
 - Designed for creating beautiful and informative statistical plots.


### Creating Statistical Plots


### Histogram

A histogram is a graphical representation of the distribution of numerical data. It consists of a series of adjacent rectangles, or bins, each representing a specific interval or range of values, and the height of each rectangle corresponds to the frequency, or count, of data points falling within that interval.

```python

import seaborn as sns
import matplotlib.pyplot as plt

# Sample data
data = sns.load_dataset("iris")

# Create a histogram
sns.histplot(data=data, x="sepal_length", kde=True)

# Adding labels and a title
plt.xlabel("Sepal Length")
plt.ylabel("Frequency")
plt.title("Histogram of Sepal Length")

# Show the plot
plt.show()
```

### Box Plot

A box plot is a method for graphically depicting groups of numerical data through their quartiles, providing a visual summary of the distribution, outliers, and variability.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data
data = sns.load_dataset("tips")

# Create a box plot
sns.boxplot(data=data, x="day", y="total_bill", palette="Set2")

# Adding labels and a title
plt.xlabel("Day")
plt.ylabel("Total Bill")
plt.title("Box Plot of Total Bill by Day")

# Show the plot
plt.show()
```

### Violin Plot

A violin plot is a method of plotting numeric data and can be considered a combination of a box plot and a kernel density plot. It displays the distribution of the data across different levels of a categorical variable. 

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data
data = sns.load_dataset("tips")

# Create a violin plot
sns.violinplot(data=data, x="day", y="total_bill", palette="pastel")

# Adding labels and a title
plt.xlabel("Day")
plt.ylabel("Total Bill")
plt.title("Violin Plot of Total Bill by Day")

# Show the plot
plt.show()
```

### Pair plot

A pair plot, also known as a scatterplot matrix, is a grid of scatterplots that displays the pairwise relationships between variables in a dataset. Each scatterplot in the grid represents the relationship between two variables, allowing for visual examination of potential correlations and patterns across multiple variables simultaneously. It's a useful tool for exploring the relationships between multiple variables in a dataset.







```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load a sample dataset
data = sns.load_dataset("iris")

# Create a pair plot
sns.pairplot(data, hue="species", markers=["o", "s", "D"])

# Show the plot
plt.show()
```

### Heat Map

A heatmap is a graphical representation of data where values in a matrix are represented as colors. It's particularly useful for visualizing relationships between two variables in a dataset. In a heatmap, each cell in the matrix is assigned a color based on its value. The colors typically range from dark (representing low values) to light (representing high values), allowing patterns and trends to be easily identified.

```python
import seaborn as sns
import matplotlib.pyplot as plt
# Load the "tips" dataset
data = sns.load_dataset("tips")
# Select relevant numeric columns
numeric_data = data[['total_bill', 'tip', 'size']]
# Calculate the correlation matrix
correlation_matrix = numeric_data.corr()
# Create a heat map
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm")
# Show the plot
plt.show()
```
